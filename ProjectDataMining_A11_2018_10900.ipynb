```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn
```

```python
dataset = pd.read_excel('dataset_tanah.xlsx')
x = dataset.iloc[:, :-1].values
y = dataset.iloc[:, 1].values
```

```python
dataset.keys()
```

```python
dataset.shape
```

```python
dataku = pd.DataFrame(dataset)
dataku.head()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(x_train, y_train)
```

```python
y_pred = regressor.predict(x_test)
```

```python
plt.scatter(dataku.LuasBangunan, dataku.HargaRumah)
plt.xlabel("Luas Bangunan")
plt.xlabel("Harga Rumah")
plt.title("Grafik LuasBangunan  vs HargaRumah")
plt.show
```

```python
plt.figure(figsize=(10,8))
plt.scatter(x_train, y_train, color = 'blue')
plt.plot(x_train, regressor.predict(x_train), color = 'red')
plt.title('Harga Rumah Terhadap Luas Tanah (Training Set)')
plt.xlabel('Luas Tanah')
plt.ylabel('Harga Rumah')
plt.show()
```

```python
plt.figure(figsize=(10,8))
plt.scatter(x_test, y_test, color = 'blue')
plt.plot(x_test, regressor.predict(x_test), color = 'red')
plt.title('Harga Rumah Terhadap Luas Tanah (Training Set)')
plt.xlabel('Luas Tanah')
plt.ylabel('Harga Rumah')
plt.show()
```
